```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country_wise_df = pd.read_csv('C:/Users/suman/Downloads/extracted_files/country_wise_latest.csv')
```

```python
country_wise_df.head()
```

```python
country_wise_df.fillna(0, inplace=True)
```

```python
total_confirmed = country_wise_df['Confirmed'].sum()
total_deaths = country_wise_df['Deaths'].sum()
total_recovered = country_wise_df['Recovered'].sum()

print("Global Summary:")
print(f"Total Confirmed Cases: {total_confirmed}")
print(f"Total Deaths: {total_deaths}")
print(f"Total Recovered: {total_recovered}")
```

```python
country_wise_df.info()
```

# BARPLOT

```python
top_10_countries = country_wise_df.nlargest(10, 'Confirmed')[['Country/Region', 'Confirmed', 'Deaths', 'Recovered']]

plt.figure(figsize=(12, 6))
sns.barplot(x='Confirmed', y='Country/Region', data=top_10_countries, palette="viridis")
plt.title("Top 10 Countries by Confirmed COVID-19 Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Country")
plt.show()
```

# PIE CHART FOR COVID DEATH CASES

```python
top_countries_death = country_wise_df.groupby("Country/Region")["Deaths"].max().nlargest(10)
plt.figure(figsize=(8,8))
pie_chart = plt.pie(top_countries_death,labels=top_countries_death.index,autopct ="%0.2f%%", startangle =140)
plt.title("Top 10 Countries with the Maximum Number of COVID-19 Death Cases")
plt.axis("equal")
plt.show()
```

```python
  
state_cases = usa_country_df.groupby('Province_State')['Confirmed'].sum()
plt.figure(figsize=(10, 10))  # Set the figure size
plt.pie(state_cases, labels=state_cases.index, autopct='%1.1f%%', startangle=140)

plt.title('Distribution of Confirmed COVID-19 Cases by State in the USA')
plt.axis('equal')  
plt.show()
```

# top n

```python
state_cases = usa_country_df.groupby('Province_State')['Confirmed'].sum()

top_n = 5
top_states = state_cases.nlargest(top_n)
others_cases = state_cases.nsmallest(len(state_cases) - top_n).sum()
if others_cases > 0:  
    top_states['Others'] = others_cases
plt.figure(figsize=(10, 10))  
plt.pie(top_states, labels=top_states.index, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Distribution of Confirmed COVID-19 Cases by State in the USA')

# Show the chart
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```
